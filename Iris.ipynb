```python
#importing all nessary libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

Loading Dataset

```python
#columns = ['Sepal Length', 'Sepal width', 'Petal Length','Petal Width', 'Class_labels']
#loadig of iris dataset

df=pd.read_csv('iris_csv.csv')
df.head(150)
```

Visualization ofDataset

```python
df.describe()
```

```python
sns.pairplot(df, hue='class')
```

Seperating input columns and output columns

```python
#seperate features and target
data=df.values

x=data[:,0:4]
y=data[:,4]

#print(x)
print(y)
```

Splittig the data into training and testing

```python
#Split the data to train and test dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

#print(x_train)
#print(y_train)
#print(x_test)
```

Model 1: Support Vector machine algorithm

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(x_train, y_train)
```

```python
prediction1 = model_svc.predict(x_test)

#calculate accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```

Model 2: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(x_train, y_train)
```

```python
prediction2=model_LR.predict(x_test)

#calculating the accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```
